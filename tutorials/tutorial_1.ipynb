# Mearuring the CO emitting layer in HD163296

In this notebook we will walk through how we can extract the emission surface using CO layers. For this example we use the DSHARP ^{12}CO J=2-1 data for HD 153296, it can be downloaded from: https://almascience.eso.org/almadata/lp/DSHARP/images/HD163296_CO.fits

We first need to initialise the cube using the casa_cube package, available at: https://github.com/cpinte/casa_cube

```python
cube_handle = casa_cube.Cube('~/HD163296_CO.fits')
```

We define the disc geometry, including any offset from the center.

```python
x0 = -2.8e-3 # arcsec
y0 = 7.7e-3  # arcsec
inc = 46.7   # deg
PA = 133.33   # deg
v_syst = 5.8 #km/s
```

Then assuming a cut off sigma we extract the emission surface.

```python
surf = cl.Surface(cube=cube_handle, PA=PA, inc= inc, v_syst=v_syst, sigma=10, dRA = x0, dDec = y0)
```

A quick way to see your extracted surface, velocity and brightness temperature is with plot_surfaces.

```python
surf.plot_surfaces()
```

You can also visulaise the peaks that were found.

```python
surf.plot_channels()
```
